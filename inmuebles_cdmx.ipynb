##Análisis exploratorio de los datos

###Valores promedio de alquiler por tipo de inmueble

```python
datos.head()
```

```python
datos["Valor"].mean() #promedio de la columna Valor
```

```python
datos.groupby("Tipo").mean() #promedio de la columna Valor por tipo de inmueble
#da jugo porque la columna colonia es del tipo string
```

```python
datos.groupby("Tipo").mean(numeric_only=True) #promedio de la columna Valor por tipo de inmueble
```

```python
valor_promedio_tipo_serie = datos.groupby("Tipo")["Valor"].mean() #promedio de la columna Valor por tipo de inmueble
print(type(valor_promedio_tipo_serie)) #es una serie
valor_promedio_tipo_serie
```

```python
valor_promedio_tipo_df = datos.groupby("Tipo")[["Valor"]].mean().sort_values("Valor") #promedio de la columna Valor por tipo de inmueble como datafra,
#ordenado de menor a mayor
print(type(valor_promedio_tipo_df)) #es un dataframe
valor_promedio_tipo_df
```

```python
valor_promedio_tipo_df.plot.bar(title="Valor promedio por tipo de inmueble")
```

```python
valor_promedio_tipo_df.plot(kind="barh", figsize=(12,8), color="purple", title="Valor promedio por tipo de inmueble")
```
